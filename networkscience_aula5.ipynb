```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Centralidade de grau

Rede n√£o direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

```python
nx.degree_centrality(G)
```

Rede direcionada

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

```python
nx.degree_centrality(H)
```

```python
nx.in_degree_centrality(H)
```

```python
nx.out_degree_centrality(H)
```
