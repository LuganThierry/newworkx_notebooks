```python
import networkx as nx
import numpy as np
```

```python
## dados de entrada: tabela de distâncias
paths = np.array([
    [0, 18, 32, 0, 0, 0],
    [18, 0, 28, 12, 0, 0],
    [32, 28, 0, 17, 4, 17],
    [0, 12, 17, 0, 0, 32],
    [0, 0, 4, 0, 0, 11],
    [0, 0, 17, 32, 11, 0]
])
```

```python
# dados de entrada: nós inicial e final
initial_node = 0
final_node = 5
```

```python
# iniciar grafo
graph = nx.from_numpy_array(paths, create_using=nx.DiGraph)
```

```python
shortest_path = nx.dijkstra_path(graph, initial_node, final_node)
shortest_path
```

```python
shortest_path_edges = [(shortest_path[i], shortest_path[i + 1]) for i in range(len(shortest_path) - 1)]
shortest_path_edges
```

```python
# resultado
edges_colors = []

for edge in graph.edges():
    if edge in shortest_path_edges:
        edges_colors.append('red')
    else:
        edges_colors.append('black')

edges_colors
```

```python
pos = nx.spring_layout(graph)
labels = nx.get_edge_attributes(graph, 'weight')
```

```python
nx.draw(graph, pos, with_labels=True, node_size=700, node_color="skyblue", edge_color=edges_colors, font_size=10)
nx.draw_networkx_edge_labels(graph, pos=pos, edge_labels=labels)
```
