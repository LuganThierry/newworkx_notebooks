```python
import networkx as nx
```

```python
c = nx.Graph()
print(c)
```

Adicionando nós

```python
c.add_node(1)
print(c)
```

```python
c.add_nodes_from([2,3])
print(c)
```

```python
c.add_nodes_from(range(12))
print(c)
```

Adicionando arestas

```python
c.add_edge(0,1)
print(c)
```

```python
c.add_edges_from([
    (1,2),
    (2,3),
    (1,3)
])
print(c)
```

Removendo nós e arestas

```python
c.remove_nodes_from(range(5))
print(c)
```

```python
c.remove_edges_from([
    (2,3),
    (1,3)
])
print(c)
```

```python
c.remove_node(8)
print(c)
```

```python
c.remove_edge(1,2)
print(c)
```

```python
c.clear()
print(c)
```

Lista de arestas

```python
edges = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'D')
]

h = nx.Graph(edges)
print(h)
```

Adjacências

```python
adjacency_dict = {
    'A': ['B', 'C'],
    'B': ['D']
}

p = nx.Graph(adjacency_dict)
print(p)
```

```python
adjacency_dict = {
    'A': ['B', 'C'],
    'B': ['D']
}

pdi = nx.DiGraph(adjacency_dict)
print(p)
```

Visualização de redes

```python
import matplotlib.pyplot as plt
```

```python
nx.draw_networkx(p)
```

```python
nx.draw_networkx(p, font_weight='bold', style='---', node_color='yellow')
```

```python
nx.draw_networkx(pdi, font_weight='bold', style='--', node_color='green')
```

```python
edges = [
    ('A', 'B', {'weight': 1}),
    ('B', 'C', {'weight': 1}),
    ('A', 'C', {'weight': 2}),
    ('E', 'F', {'weight': 3}),
]

t = nx.Graph(edges)
print(t)
```

```python
pos = nx.spring_layout(t)
labels = nx.get_edge_attributes(t, 'weight')
nx.draw_networkx(t, pos)
nx.draw_networkx_edge_labels(h, pos, edge_labels=labels)
```

Layouts


```python
nx.draw_circular(t, with_labels=True)
```

```python
nx.draw_kamada_kawai(t, with_labels=True)
```

```python
nx.draw_shell(t, with_labels=True)
```
