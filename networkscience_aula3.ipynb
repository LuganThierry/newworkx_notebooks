```python
import networkx as nx
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
```

Funções de grau e rank

```python
def plot_degree(graph):
    degree_list = [degree for node, degree in graph.degree]
    degree_series = pd.Series(degree_list)
    degree_series.hist()
    plt.xlabel('Grau')
    plt.ylabel('Frequência')
    plt.title('Distribuição de graus')
```

```python
def plot_rank(graph, log=False, weight=''):
    degree_list = [degree for node, degree in graph.degree]
    y = sorted(degree_list, reverse=True)
    x = range(len(y))
    plt.plot(x, y)
    plt.xlabel('Posição (rank)')
    plt.ylabel('Grau')
    if log:
        plt.loglog()
```

Clube de karatê

```python
K = nx.karate_club_graph()
print(K)
```

```python
nx.draw_circular(K, with_labels=True)
```

Mundo pequeno

```python
nx.density(K)
```

```python
nx.average_shortest_path_length(K)
```

```python
def get_distance(graph):
    distance_list = []
    for source, dist in nx.shortest_path_length(graph):
        distance_list += dist.values()
    return pd.Series(distance_list)
```

```python
distance_list = get_distance(K)
distance_list.describe()
```

Meus amigos são amigos

```python
nx.average_clustering(K)
```

Desigualdade de rede

```python
plot_degree(K)
```

Tudo está conectado

```python
nx.is_connected(K)
```
